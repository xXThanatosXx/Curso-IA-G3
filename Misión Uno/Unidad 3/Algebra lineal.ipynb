## Vectores

1. **Vectores geométricos.** Este ejemplo de un vector puede ser familiar de las matemáticas y física de la escuela secundaria. Los vectores geométricos - ver Figura 2.1(a) - son segmentos dirigidos, los cuales pueden ser dibujados (al menos en dos dimensiones). Dos vectores geométricos $\vec{x}$, $\vec{y}$ pueden ser sumados, de tal forma que $\vec{x} + \vec{y} = \vec{z}$ es otro vector geométrico. Además, la multiplicación por un escalar $\lambda \vec{x}$, $\lambda \in \mathbb{R}$, también es un vector geométrico. De hecho, es el vector original escalado por $\lambda$. Por lo tanto, los vectores geométricos son instancias de los conceptos de vector introducidos previamente. Interpretar los vectores como vectores geométricos nos permite usar nuestra intuición acerca de la dirección y magnitud para razonar sobre operaciones matemáticas.



```python
import matplotlib.pyplot as plt
import numpy as np



# Define the system of equations
# 4x1 + 4x2 = 5 --> x2 = (5 - 4x1)/4
# 2x1 - 4x2 = 1 --> x2 = (2x1 - 1)/4

# Create the range of x1 values
x1 = np.linspace(-10, 10, 400)

# Calculate x2 values for each equation
x2_eq1 = (5 - 4*x1)/4
x2_eq2 = (2*x1 - 1)/4

# Plot the lines
plt.figure(figsize=(5, 4))

# Equation 1: 4x1 + 4x2 = 5
plt.plot(x1, x2_eq1, label='4x1 + 4x2 = 5', color='blue')

# Equation 2: 2x1 - 4x2 = 1
plt.plot(x1, x2_eq2, label='2x1 - 4x2 = 1', color='red')

# Add labels and title
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('System of Equations')

# Find the intersection point by solving the system of equations
A = np.array([[4, 4], [2, -4]])
b = np.array([5, 1])
intersection = np.linalg.solve(A, b)

# Plot the intersection point
plt.plot(intersection[0], intersection[1], 'ko') # 'ko' is for black circle

# Adding legend
plt.legend()

# Show grid
plt.grid(True)

# Set x and y axis limits
plt.xlim(-2, 5)
plt.ylim(-2, 5)

# Save the figure
#plt.savefig('system_of_equations.png')
# Display the plot
plt.show()
```

## Matrices: Inversa

```python
import numpy as np

# Defining the matrix A
A = np.array([[1, 0, 2, 0],
              [1, 1, 0, 0],
              [1, 2, 0, 1],
              [1, 1, 1, 1]])


A_pseudo_inverse = np.linalg.pinv(A)

A_pseudo_inverse = np.round(A_pseudo_inverse)

print(A_pseudo_inverse)
```
