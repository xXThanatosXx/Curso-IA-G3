## Vectores

1. **Vectores geométricos.** Un vector puede ser familiar de las matemáticas y física de la escuela secundaria. Los vectores geométricos - ver Figura (a) y (b), son segmentos dirigidos, los cuales pueden ser dibujados (al menos en dos dimensiones). Dos vectores geométricos $\vec{x}$, $\vec{y}$ pueden ser sumados, de tal forma que $\vec{x} + \vec{y} = \vec{z}$ es otro vector geométrico. Además, la multiplicación por un escalar $\lambda \vec{x}$, $\lambda \in \mathbb{R}$, también es un vector geométrico. De hecho, es el vector original escalado por $\lambda$. Por lo tanto, los vectores geométricos son instancias de los conceptos de vector introducidos previamente. Interpretar los vectores como vectores geométricos nos permite usar nuestra intuición acerca de la dirección y magnitud para razonar sobre operaciones matemáticas.



En inteligencia artificial, los vectores son a menudo utilizados para representar datos de entrada, pesos en modelos de aprendizaje automático, o características de los objetos en un conjunto de datos. Utilizando Python, podemos demostrar el uso de vectores en operaciones típicas de IA como la multiplicación de vectores por escalares, suma de vectores, producto punto y la manipulación de vectores con bibliotecas como NumPy.

```python
import numpy as np

# Crear dos vectores
vector_a = np.array([2, 3])
vector_b = np.array([0, 1])

# Suma de vectores
suma = vector_a + vector_b

# Producto escalar (dot product)
#Suma de los productos de los correspondientes elementos de los vectores, una medida de similitud.
producto_punto = np.dot(vector_a, vector_b)

# Multiplicación por un escalar
escalar = 3
vector_escalar = vector_a * escalar

# Distancia euclidiana entre dos vectores
#dos puntos en el espacio n-dimensional, útil para muchos algoritmos de clasificación y clustering.
distancia = np.linalg.norm(vector_a - vector_b)

# Normalización de un vector
norma = np.linalg.norm(vector_a)
vector_normalizado = vector_a / norma

print("Suma de vectores:", suma)
print("Producto escalar:", producto_punto)
print("Vector multiplicado por escalar:", vector_escalar)
print("Distancia Euclidiana:", distancia)
print("Vector normalizado:", vector_normalizado)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Dibujar la distancia euclidiana entre vectores
# Creando dos vectores
vector_a = np.array([2, 3])
vector_b = np.array([0, 1])
# Calculando la distancia euclidiana

distancia = np.linalg.norm(vector_a - vector_b)

# Configurando el gráfico
plt.figure(figsize=(6, 6))
# Dibujando los vectores
plt.quiver(0, 0, vector_a[0], vector_a[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vector A')
plt.quiver(0, 0, vector_b[0], vector_b[1], angles='xy', scale_units='xy', scale=1, color='b', label='Vector B')

# Dibujando la distancia como una línea punteada
plt.plot([vector_a[0], vector_b[0]], [vector_a[1], vector_b[1]], 'g--', label='Distancia Euclidiana')

# Configuración adicional del gráfico
plt.xlim(-1, 4)
plt.ylim(-1, 4)
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.title('Distancia Euclidiana entre Vectores')
plt.grid(True)
plt.show()

print(f"La distancia euclidiana es: {distancia:.2f}")
```

```python
# Normalizando el vector A
vector_a_norm = vector_a / np.linalg.norm(vector_a)
# Configurando el gráfico
plt.figure(figsize=(6, 6))
# Dibujando el vector original A
plt.quiver(0, 0, vector_a[0], vector_a[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vector A')
# Dibujando el vector normalizado A
plt.quiver(0, 0, vector_a_norm[0], vector_a_norm[1], angles='xy', scale_units='xy', scale=1, color='g', label='Vector A Normalizado')

# Configuración adicional del gráfico
plt.xlim(-1, 3)
plt.ylim(-1, 3)
plt.gca().set_aspect('equal', adjustable='box')
plt.legend()
plt.title('Vector Normalizado')
plt.grid(True)
plt.show()

print(f"Vector A normalizado: {vector_a_norm}")
```

## Matrices

- Transpuesta de una matriz:
La transpuesta de una matriz, que es el intercambio de filas por columnas, se utiliza frecuentemente en algoritmos de aprendizaje automático para ajustar el formato de los datos o en operaciones de redes neuronales para el cálculo de gradientes.

- Multiplicación de una matriz por un vector:
Esta operación es esencial en el procesamiento de señales y en redes neuronales, donde el vector puede representar un conjunto de pesos que se aplica a las entradas (representadas por la matriz) para obtener salidas o características.

- Inversa de una matriz:
La inversa de una matriz se utiliza en métodos de optimización, como en la solución de sistemas de ecuaciones lineales, que pueden surgir, por ejemplo, en la regresión lineal y otros métodos de aprendizaje supervisado.

- Autovalores y autovectores:
En técnicas de reducción de dimensionalidad como el Análisis de Componentes Principales (PCA), los autovalores y autovectores de la matriz de covarianza se utilizan para identificar las direcciones de mayor varianza en los datos, lo que permite reducir la cantidad de dimensiones preservando la mayor cantidad de información posible.

```python
import numpy as np

# Creación de una matriz de ejemplo
matriz = np.array([[2, 3], 
                   [1, 5], 
                   [0, 6]])

# Operaciones comunes con matrices en IA

# Transpuesta de la matriz
transpuesta = matriz.T

# Multiplicación de matriz por un vector
vector = np.array([1, 2])
producto_mv = np.dot(matriz, vector)

# Multiplicación de dos matrices (por ejemplo, en redes neuronales)
matriz_b = np.array([[1, 2, 3], 
                     [4, 5, 6]])
producto_mm = np.dot(matriz, matriz_b)

# Inversa de una matriz cuadrada (por ejemplo, en la optimización de mínimos cuadrados)
matriz_cuadrada = np.array([[1, 2], 
                            [3, 4]])
inversa = np.linalg.inv(matriz_cuadrada)

# Autovalores y autovectores (por ejemplo, en PCA para reducción de dimensionalidad)
autovalores, autovectores = np.linalg.eig(matriz_cuadrada)

# Explicación de las operaciones

explicaciones = {
    'transpuesta': transpuesta,
    'producto_mv': producto_mv,
    'producto_mm': producto_mm,
    'inversa': inversa,
    'autovalores': autovalores,
    'autovectores': autovectores
}

explicaciones

```

### Transpuesta de una matriz

```python
import numpy as np
import matplotlib.pyplot as plt

# Matriz original
matriz = np.array([[2, 3], 
                   [1, 5], 
                   [0, 6]])

# Transpuesta de la matriz
transpuesta = matriz.T


# se pueden visualizar los elementos de la matriz y su transpuesta como imágenes.
plt.matshow(matriz)
plt.title('Matriz Original')
plt.show()

plt.matshow(transpuesta)
plt.title('Transpuesta de la Matriz')
plt.show()
```

### Multiplicación de la matriz por un vector

```python
import numpy as np
import matplotlib.pyplot as plt

# Matriz original
matriz = np.array([[2, 3], 
                   [1, 5], 
                   [0, 6]])
# Vector
vector = np.array([1, 2])

# Multiplicación de la matriz por un vector
producto_mv = np.dot(matriz, vector)
print(producto_mv)
```

### Producto matricial

```python
# Segunda matriz para la multiplicación
matriz_b = np.array([[1, 2, 3], [4, 5, 6]])

# Multiplicación de matrices
producto_mm = np.dot(matriz, matriz_b)

# Esta operación también resulta en una matriz, que no se puede 'graficar' directamente.
# Podemos visualizar la matriz resultante como una imagen.
plt.matshow(producto_mm)
plt.title('Producto de dos Matrices')
plt.show()
```

### Inversa de una matriz

```python
# Matriz cuadrada
matriz_cuadrada = np.array([[1, 2], [3, 4]])

# Cálculo de la inversa
inversa = np.linalg.inv(matriz_cuadrada)

plt.matshow(matriz_cuadrada)
plt.title('Matriz Cuadrada')
plt.show()

# Visualización de la matriz inversa
plt.matshow(inversa)
plt.title('Inversa de la Matriz Cuadrada')
plt.show()
```

```python

# Definimos una matriz cuadrada
matriz_cuadrada = np.array([[1, 2],
                            [3, 4]])

# Calculamos los autovalores y autovectores
autovalores, autovectores = np.linalg.eig(matriz_cuadrada)

# Creamos la figura para el gráfico
plt.figure()

# Origen de los autovectores
origin = np.zeros((2, autovectores.shape[1]))

# Graficamos los autovectores
plt.quiver(*origin, autovectores[0, :], autovectores[1, :], color=['r', 'b'], scale=21)

# Establecemos los límites del gráfico
plt.xlim(-0.25, 0.25)
plt.ylim(-0.25, 0.25)

# Añadimos una cuadrícula, título y mostramos el gráfico
plt.grid(True)
plt.title('Autovectores de la Matriz Cuadrada')
plt.show()

# Mostramos los autovalores y autovectores
print(f"autovalores:",autovalores)
print(f"autovectores:",autovectores)
```

## TENSOR

Un tensor es una generalización de vectores y matrices a un número arbitrario de dimensiones (también conocido como un array n-dimensional). En el contexto de la inteligencia artificial y, más específicamente, en el aprendizaje automático y profundo, los tensores se utilizan como la estructura de datos principal para almacenar y procesar datos.

- Datos de Entrada: En muchos modelos de IA, especialmente en las redes neuronales, los tensores son utilizados para representar los datos de entrada. Por ejemplo, una imagen puede ser representada como un tensor de 3 dimensiones con dimensiones correspondientes a la altura, anchura y canales de color.

- Pesos de Modelo: Los pesos en una red neuronal se organizan comúnmente en tensores, que se actualizan durante el proceso de entrenamiento.

- Operaciones de Tensor: La mayoría de las operaciones en redes neuronales son operaciones tensoriales que incluyen suma de tensores, multiplicación de tensores, convoluciones (especialmente en redes neuronales convolucionales), etc.

- Gradientes: Durante la retropropagación, se calculan los gradientes de los tensores de peso para actualizar los pesos del modelo.

- Representaciones de Alto Nivel: Los tensores son utilizados para representar características de alto nivel en el procesamiento de lenguaje natural y en visión por computadora, donde las dimensiones pueden representar diferentes aspectos de los datos, como las palabras en un texto o los fragmentos de una imagen.

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

# Ejemplo de tensor en TensorFlow
tensor = tf.constant([[1, 2, 3], 
                      [4, 5, 6]])

# Operaciones básicas con tensores en IA

# Suma de un escalar a un tensor
tensor_mas_escalar = tensor + 2

# Producto elemento a elemento de dos tensores
tensor_b = tf.constant([[1, 1, 1], 
                        [1, 1, 1]])
producto_elemento = tf.multiply(tensor, tensor_b)

# Redimensionar un tensor
tensor_reshaped = tf.reshape(tensor, [3, 2])

# Ejemplo de una operación más compleja: convolución 2D
# Crear un tensor 4D (imagen con un solo canal y un lote de una sola imagen)
imagen = tf.random.normal([1, 10, 10, 1])
# Crear un filtro tensor 4D (filtro para una sola entrada y salida de canal)
filtro = tf.random.normal([3, 3, 1, 1])
# Aplicar la convolución
imagen_convolucionada = tf.nn.conv2d(imagen, filtro, strides=[1, 1, 1, 1], padding='SAME')

# Explicación de las operaciones

explicaciones = {
    'tensor_original': tensor,
    'tensor_mas_escalar': tensor_mas_escalar,
    'producto_elemento': producto_elemento,
    'tensor_reshaped': tensor_reshaped,
    'imagen_convolucionada': imagen_convolucionada[0, :, :, 0]
}

# Ejemplo de visualización de un tensor como imagen
plt.imshow(imagen[0, :, :, 0], cmap='gray')
plt.title('Ejemplo de Tensor como Imagen')
plt.show()

explicaciones

```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# Creando un tensor 3D como una lista de matrices 2D
tensor_3d = np.array([np.random.rand(4, 4) for _ in range(10)])

# Configurar la figura y los ejes para la animación
fig, ax = plt.subplots(figsize=(5, 5))

# Función de animación que se llama secuencialmente
def animate(i):
    ax.clear()  # Limpiar el gráfico anterior
    ax.matshow(tensor_3d[i], cmap='viridis')  # Mostrar la matriz
    ax.set_title(f'Slice {i}')  # Establecer el título para mostrar el índice de tiempo

# Animación usando FuncAnimation sin blitting
anim = FuncAnimation(fig, animate, frames=len(tensor_3d), interval=500)

# Mostrar la animación en Jupyter Notebook
HTML(anim.to_jshtml())

```
